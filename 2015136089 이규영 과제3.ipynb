```python
def addall(one) :
    return sum([i for i in one])

lst = [1,2,3,4,5,6,7,8,9]
print(addall(lst))

def addallodd(two) : 
    return sum([i for i in two if(i%2 == 1)])

print(addallodd(lst))
```

```python
L1 = [1, 2, 3]
L2 = [4, 5, 6]
d = {'low':L1, 'high':L2}
e = d
f = d.copy()
print(d)
print(e)
print(f)
print()
d['low'] = [10, 20, 30]
d['high'][1] = 500
print(d)
print(e)
print(f)
```

+ 1) d는 L1과 L2를 사전형태로 가리키고 있으며 e는 그런 d를 가리키고 있다. 그렇기에 값을 복사한 것이 아닌 포인터 형태임으로 둘은 항상 같은 값을 반환한다.
+ 2)  = 은 주소를 복사하는 개념일시에 copy는 주소가 아닌 값을 복사한다. 그렇기에 d의 low가 가리키는 [10,20,30]은 영향을 미치지않고 기존의 복사한 값인 [1,2,3]을 반환한다. 하지만 high[1]을 통해 값 자체를 500으로 변화시켰기 때문에 값이 복사되어 [4,500,6]을 반환한다.

```python
def cmp(a):
    return a[1]

d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
blank = []
for (i,j) in d.items():
    blank.append((i,j))
blank.sort()
print (blank)
blank.sort(reverse=True)
print(blank)
blank.sort(key = cmp)
print (blank)
blank.sort(key = cmp, reverse=True)
print(blank)
```

+ d.item은 d.key와 d.values를 둘다 합친 즉 키값과 밸류값을 모두 불러내는 함수이다.append를 통해 blank에 삽입한다.
+ sort를 통해 key를 오름차순으로 출력한다.
+ reverse는 항상 디폴트인 false를 지니고 있으나 true로 값을 변경시켜주어 오름차순이 된 사전을 내림차순으로 뒤집어 준다.
+ cmp(a)함수는 기존에 디폴트 되어있던 값을 반대로 변경해주는 함수이다.
+ key = cmp를 통해 항상 키를 반환하던 사전에서 밸류를 반환하도록 바꾸어 주고 밸류를 기준을 sort한다.
+ 위와 같이 reverse를 통해 다시 뒤집어준다.

```python
import sys
for line in sys.stdin:
    coupon,k=map(int,line.split())
    total = coupon
    while(coupon >= k):
        total = total + coupon//k
        coupon = coupon//k + coupon%k
    print (total)
```

+ import sys와 for문은 구글링을 통해서 백준내에 정답이 되도록 참고하였다.
+ 쿠폰갯수를 치킨 구매시 가능한 total에 저장한다.
+ while문을 통해 도장으로 치킨구매가 불가능할 때까지 반복시킨다.
+ 후에 치킨 쿠폰과 도장으로 구매할 수 있는 치킨수를 total에 저장한다.
+ 치킨을 주문하고 올때 가져온 도장으로 다시 치킨을 주문하고 다시 주문한 치킨에 온 도장으로 가능한 치킨 갯수를 다시 구한다.
+ 도장으로 주문이 가능할 때 까지 반복시킨다.

```python
import requests
import string

def wordfunction(abc) :
    req = requests.get(abc)
    source = req.text
    stp = string.punctuation

    text = "" #순수 텍스트 식별자
    while(1) :      
        source = source[source.find('<'):] #'<'를 찾는다
        if source.find("<!--") == 0 : source = source[source.find("-->")+3:] #주석 내용 건너뜀
        elif source.find("<script>") == 0 : source = source[source.find("</script>")+9:] #자바스크립트 내용 건너뜀   
        else :
            source = source[source.find('>')+1:] #'>'를 찾는다
            for ch in source :    #'>' 이후부터 한글자씩 가져온다
                if ch == '<' : break    # 다시 '<' 문자 등장하면 종료
                elif ch == '\t' or ch == '\n' : continue    #탭,개행 문자는 순수 텍스트에 포함x
                else : text += ch   #순수 텍스트 추가     
            if text.endswith(" ") == False : text += " " #공백으로 분리하기 위해 순수 텍스트 추출 후 뒤에 공백문자 추가
        if(source.find('<') == -1):break #'<' 문자가 더 이상 없는 경우 무한루프종료 

    # print(text)   

    list = text.split() #추출한 순수 텍스트 문자열을 공백 단위로 분리
    #print (list)

    #print (len(list)) #분리된 단어 수 출력

    for i in range(len(list)):
        list[i] = list[i].strip()
        list[i]= list[i].strip(stp)

    for i in range(len(list)):
        for j in stp:
            while( j in list[i]):
                for k in range(len(list[i])):
                    check = list[i].find(j)
                    list[i] = list[i][:check] + list[i][check+1:]
                    break


    k=0
    for i in list:
        if ( i == ''):
            k+=1
    for i in range(k):
        list.remove('')

    dic = dict()
    for i in list:
        dic[i] = 0
    for i in dic.keys():
        for j in list:
            if ( j == i):
                dic[i]+=1
    print(dic)
    
wordfunction('http://cse.koreatech.ac.kr')
print()
wordfunction('https://www.koreatech.ac.kr')
print()
wordfunction('http://www.naver.com')
print()
wordfunction('http://www.daum.net')
print()
wordfunction('http://www.nytimes.com')
```

+ 모범답안을 기준으로 작성하였습니다.(요구사항1 무시)
+ string.puncutation을 stp에 저장후 stip을 이용하여 구두문자를 모두 제거한다.(요구사항 2)
+ 공백으로 판명되는 단어들을 제거하기 위해 list내에서 stp를 포함한 것을 찾는다. 이때 list내 stp가 없을때까지 반복시키고 그 내에서 stp를 find로 찾고 check에 저장한다. check를 기준으로 앞과 그 뒤로 이어서 list에 다시 저장한다. (요구사항 3)
+ 빈 사전 dic을 생성하고 list내에 밸류값을 0으로 초기화한다. 사전적 키와 리스트 내에 같은 단어가 반복되면 밸류값을 +1 시키고 dio사전에 저장한다. 후에 단어와 빈도수를 출력한다.
